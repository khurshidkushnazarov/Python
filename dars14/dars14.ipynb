```python
def categorize(code):      # 4 talik bulsa viloyat, aks holda Tuman/Shahar deb yozsin degan funksiya
    if code == 1700 or code == 1735:
        return 'respublika'
    elif len(str(code)) == 4:
        return 'Viloyat'
    else:
        return 'Tuman/Shahar'
```

```python
demo["category"] = demo["Code"].apply(categorize)
```
