```python
d = {
    "Region":["Tashkent","Tashkent","Tashkent","Tashkent","Samarkand","Samarkand","Samarkand"],
    "City":["Chilonzor", "Yashnobod","Mirobod","Sergeli","Bulungur","Urgut","Jomboy"],
    "aholi":[5000,4000,8000,6000,2500,2000,1500]
}
df3 = pd.DataFrame(d)
df3
```

```python
df3.groupby("Region").agg({"aholi":"sum"})
```

```python
df3.groupby("Region").agg({"aholi":["sum","mean","max","min"]})
```

```python
df3.groupby("Region").agg({"aholi":["sum","mean","max","min"],"City":"count"})
```

```python
df3.groupby(["Region","City"]).agg({"aholi":"sum"})
```

```python
df3.groupby(["Region","City"]).agg({"aholi":"sum"}).reset_index()
```

```python
d = {
    "Region":["Tashkent","Tashkent","Tashkent","Tashkent","Samarkand","Samarkand","Samarkand"],
    "City":["Chilonzor", "Yashnobod","Mirobod","Sergeli","Bulungur","Urgut","Jomboy"],
    "aholi":[5000,4000,8000,6000,2500,2000,1500]
}
df3 = pd.DataFrame(d)
df3
```

```python
df4 = df3.groupby("Region").agg({"aholi":["sum","mean","max","min"],"City":"count"})
```

```python
df4.columns
```

```python
columns = []
for i in df4.columns:
    columns.append("_".join(i))
columns
```

```python
df4.columns = columns
```

```python
df4.reset_index(inplace=True)
```

```python
df4.to_csv('countries.csv',index=False)
```

```python
df4.to_excel('countries.xlsx',index=False)
```
