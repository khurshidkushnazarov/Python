```python
def add_numbers(*numbers):  # sonlar 0 ta yoki istalgancha degani--- *numbers
    print(sum(numbers))
add_numbers(10,15,20,30)
```

```python
### kupaytirish  (res uzgaruvchi)
def add_numbers(*numbers):
    res=1
    for i in numbers:
        res=res*i
    print(res)
add_numbers(10,15,20,30)
```

```python
def power_two(number):
    num = 1
    while 2**num < number:
        print(2**num)
        num=num+1

power_two(100)
```

```python
def palindrom(word):
    if word == word[::-1]:
        print('Bu palindrom')
    else:
         print('Bu palindrom emas')   
palindrom('anna')         
```

```python
### d={'blue': 2, 'red': 1, 'yellow': 1, 'green': 2}

ls = ['blue', 'red', 'yellow', 'green', 'blue', 'green']
def func(colors):
    return ls.count(colors)
d={}
for i in ls:
    d[i]=func(i)
print(d)
```

```python
ls = ['blue', 'red', 'yellow', 'green', 'blue', 'green']

d={}

for word in ls:
    if word not in d.keys():
        d[word] = 1
    else:
        d[word] = d[word] +1   
d        
```
