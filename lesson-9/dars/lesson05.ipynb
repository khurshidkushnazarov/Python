```python
year = int(input("Yilni kiriting: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} - Kabisa yili.")
else:
    print(f"{year} - oddiy yil.")
```

```python
## Undoshlarni topish
country = 'Uzbekistan'
unli = 'aioueAIOUE'
for letter in country:
    if letter not in unli:
        print(f'Undosh: {letter}')
```

```python
## Undoshlarni topish
country = 'Uzbekistan'
unli = 'aioue'
for letter in country:
    if letter.lower() not in unli:
        print(f'Undosh: {letter}')
```
